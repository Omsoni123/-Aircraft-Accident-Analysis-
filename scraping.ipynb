```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

website = 'https://www.adamchoi.co.uk/bttsresult/detailed'

driver.get(website)
```

```python
driver.quit()
```

```python
football=driver.find_element(By.XPATH, "//label[@analytics-event='All matches']")
```

```python
football.click()
```

```python
matches=driver.find_elements(By.TAG_NAME,"tr")
```

```python
date=[]
home_team=[]
Score=[]
away_team=[]

```

```python
for match in matches:
    date_text = match.find_element(By.XPATH, ".//td[1]").text
    home_text = match.find_element(By.XPATH, ".//td[2]").text
    score_text = match.find_element(By.XPATH, ".//td[3]").text
    away_text = match.find_element(By.XPATH, ".//td[4]").text

    date.append(date_text)
    home_team.append(home_text)
    Score.append(score_text)
    away_team.append(away_text)
    print(home_text)
```

```python
print("Date:", date)
```

```python
import pandas as pd
data = pd.DataFrame({'date': date, 'home_team': home_team, 'Score': Score, 'away_team': away_team})
data.to_csv('football_matches.csv', index=False)
```
